```python
%matplotlib notebook
from skmultiflow.meta.multi_output_learner import MultiOutputLearner
from skmultiflow.data.multilabel_generator import MultilabelGenerator
from lstm_stream import LSTMMask,EvaluatePrequential
```

```python
def demo(output_file=None, n_inst=None):  
    nb_nf=5
    nb_metrics=10
    batch_size = 1
    epochs = 10
    embedding_size = 128
    filters = 64
    kernel_size = 5
    lstm_output_size = nb_nf
    max_features = nb_nf*nb_metrics
    maxlen = max_features
    pool_size = 4

    # Setup the File Stream
    #stream = FileStream("music.csv", 0, 6)
    stream = MultilabelGenerator(n_samples=n_inst+1, n_features=max_features, n_targets=lstm_output_size,                                 n_labels=2, random_state=0)
    stream.prepare_for_use()
    print(stream)
    # Setup the classifier
    #classifier = MultiOutputLearner(SGDClassifier(max_iter=100))
    classifier = MultiOutputLearner(LSTMMask(max_features=max_features, maxlen=maxlen, embedding_size=embedding_size, kernel_size=kernel_size, filters=filters, pool_size=pool_size,
                 lstm_output_size=lstm_output_size, batch_size=batch_size, epochs=epochs))
    print(classifier)
    # Setup the evaluator
    evaluator = EvaluatePrequential(pretrain_size=1000, max_samples=n_inst-1000, batch_size=1, n_wait=200,
                                    max_time=1000, output_file=output_file, show_plot=True,
                                    metrics=['hamming_score','hamming_loss', 'j_index', 'exact_match'])
#n_wait : The number of samples to process between each test. 
#max_samples : The maximum number of samples to process during the evaluation.
#pretrain_size : The number of samples to use to train the model before starting the evaluation. 
#max_time : The maximum duration of the simulation (in seconds).
    # Evaluate
    evaluator.evaluate(stream=stream, model=classifier ,model_names=['RCA'])
```

```python
if __name__ == '__main__':
    demo('test_prequential_mol.csv', n_inst=50000,)
```
